# Thermal performance

Getting instant q [W] values. Considering the central three columns of panels.

The heat transfered to the fluid can be expressed by the heat balance:
$$q = c_p \dot{m}(T_{f,in} - T_{f,out}) $$

where $q$ is the heat gain in Watts, $c_p$ the specific heat of air in the duct and $\dot m$ the mass flow in the duct. The driving force is the temeperature difference between entrance and output temperatures. By considering $T_{f,in} = T_{ambient}$ and assuming that the flow inside the gap has a bulk uniform mean velocity $U_{mean}$, the instant $q$ can be calculated.

```python
dft['U_mean']= (dft.U_f1 + dft.U_f2 + dft.U_f3) / 3
dft['deltaT_air'] = dft.T_f3 - dft.T_amb_avg
dft['mdot'] = dft.U_mean * A_c * rho
dft['q'] = dft.mdot * cp * dft.deltaT_air
```

# Electrical Performance

For the same data range, the initial approach is done by getting the electrical performance ratio (PR) of the complete system:

$$\bar{PR} = \frac{Y_{f}}{Y_{ref}} = \frac{\sum P_{DC}} {    P_{STC} * n_{mod}* \frac{\sum G_{POA}}{1000}  } $$  
where $Y_{ref}$ is the reference yield and $Y_f$ is the real electrical yield of the complete system.

```python
PR_avg = (dft.P_AC.sum()) / (250 * 16 *dft.G_poa_w.sum() / 1000)
dft['Y_ref'] = (dft.loc[:,'G_poa_w'] * 250 * 16) / 1000             #instantaneous system yield
```

```python
plt.plot(dft.index, dft[['Y_ref', 'P_AC']])
plt.legend(['Y_ref', 'P_AC'], loc=2)   # 1 =bestlocation, 2 =upperright, 3 =upperleft, etc
plt.ylabel('$[W]$', fontsize=15)
plt.xlabel('time'); plt.legend(loc =2); plt.grid() ; plt.title("Instant $Y_{ref}$ and $Y_{f}$ ")
plt.gcf().set_size_inches(12,3)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%H:%M')) #getting desired format for x axis
```
